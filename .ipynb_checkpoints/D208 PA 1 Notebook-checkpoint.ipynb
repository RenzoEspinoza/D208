# Part IV: Model Comparison and Analysis

```python
predictors = ['Bandwidth_GB_Year', 'Device_protection', 'Online_backup', 'Two_year',
              'Rural', 'Female', 'Yearly_equip_failure', 'Urban', 'Month_to_month',
              'Nonbinary','Multiple', 'No_internet', 'Techie']
```

```python
sample = df.sample(2500, random_state=1)
X = sample[predictors]
y = sample['Tenure']
X = sm.add_constant(X)
init_model = sm.OLS(y, X).fit()
init_model.summary()
```

    Simpler yet are forward selection and backward selection. In forward selection, you start with no predictors and add them one by one, at each step adding the predictor that has the largest contribution to R2, and stopping when the contribution is no longer statistically significant. In backward selection, or backward elimination, you start with the full model and take away predictors that are not statistically significant until you are left with a model in which all predictors are statistically significant.


p. 159 "Practical Statistics for Data Scientists", 2nd edition

```python
def forward_select(predictors):
    models_df = pd.DataFrame()
    models_df['vars'] = pd.Series(dtype='object')
    models_df['rsquared_adj'] = pd.Series(dtype='float')
    x = pd.DataFrame()
    best_model = None
    for i in range(len(predictors)):
        x = pd.DataFrame()
        if i > 0:
            best_model = models_df.iloc[0, :]
            x[best_model.vars] =sample[best_model.vars]
            models_df = models_df[0:0]
        
        for j in range(len(predictors)):
            predictor = predictors[j]
            x[predictor] =sample[predictor]
            x = sm.add_constant(x)
            result = sm.OLS(y, x).fit()
            x = x.drop(['const'], axis=1)
            entry = pd.DataFrame([{'vars': x.columns.values, 'rsquared_adj': result.rsquared_adj}])
            models_df = pd.concat([models_df, entry], axis=0, ignore_index=True)
            x = x.drop([predictor], axis=1)
            
        models_df = models_df.sort_values(by=['rsquared_adj'], ascending=False)
        display(models_df)
        curr_model = models_df.iloc[0, :]
        predictor = curr_model.vars[-1]
        if(i > 0):
            diff = curr_model.rsquared_adj - best_model.rsquared_adj

            if(diff <= 0.01):
                print(f"The predictor {predictor}'s contribution to the coefficient of determination ({diff}) was not statistically significant")
                break
        
        print(f"The predictor with the largest contribution to the coefficient of determination in this iteration is {predictor}, bringing R2 up to {curr_model.rsquared_adj}")
        predictors.remove(predictor)

    return best_model

result = forward_select(predictors)

display(result.vars.tolist())
display(result.rsquared_adj)
```

### Because the PA requires both categorical and continuous variables in the reduced model, I will include No_internet

## Comparison

```python
init_model.summary()
```

```python
reduced_predictors = ['Bandwidth_GB_Year', 'No_internet']
X = sample[reduced_predictors]
X = sm.add_constant(X)
reduced_model = sm.OLS(y, X).fit()
reduced_model.summary()
```

```python
print(f"Initial model residual standard error: {np.sqrt(init_model.mse_resid)}")
print(f"Reduced model residual standard error: {np.sqrt(reduced_model.mse_resid)}")
```

## White's Test to check for heteroscedasticity

```python
whites_test = het_white(init_model.resid, init_model.model.exog)
test_labels = ['Test Statistic', 'Test Statistic p-value', 'F-Statistic', 'F-Test p-value']
print('Initial Model: ')
print(dict(zip(test_labels, whites_test)))
```

```python
whites_test = het_white(reduced_model.resid, reduced_model.model.exog)
print('Reduced Model:')
print(dict(zip(test_labels, whites_test)))
```

### Since the Test Statistic p-value for the reduced model is below 0.05, we can reject the null hypothesis, meaning that heteroscedasticity is present

## Durbin Watson Test for autocorrelation

```python
print('Initial Model:')
durbin_watson(init_model.resid)
```

```python
print('Reduced Model:')
durbin_watson(reduced_model.resid)
```

### Since the test statistics is within the range of 1.5 and 2.5, autocorrelation is not present
